## Scraping song lyrics

```python
#shoutout to James Pecore for this code!
lyrics_list = []
genius.verbose = False # keeps us from having to see every song printed out as its lyrics are searched (saves time)

start_time = time.time()
for i in billboard_df.index:
    artist = billboard_df.loc[i,'artist']
    print(artist)
    track_name = billboard_df.loc[i,'track_name']
    print(track_name)
    try:
        song = genius.search_song(track_name, artist)
        lyrics = song.lyrics
    except:
        lyrics = ''
    lyrics_list.append(lyrics)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
billboard_df['lyrics'] = lyrics_list
```
